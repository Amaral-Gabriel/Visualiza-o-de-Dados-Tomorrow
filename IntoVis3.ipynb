```python
import plotly.express as px
import pandas as pd
```

```python
df = px.data.iris()
fig = px.scatter_matrix(df, dimensions=["sepal_length","sepal_width","petal_length","petal_width"], color="species")
fig.show()
```

```python
fig = px.parallel_coordinates(df, dimensions=["sepal_length","sepal_width","petal_length","petal_width"], color="species_id")
fig.show()
```

```python
from sklearn.manifold import TSNE

df
features = df.loc[:, :'petal_width']

tsne = TSNE(n_components=3, random_state=0)
projections = tsne.fit_transform(features,)

projections

fig = px.scatter_3d(projections, x=0, y=1, z=2, color=df.species, labels={"color": "spcies"})
fig.show()
```
